```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv('cost_revenue_clean.csv')
```

```python
data.describe()
```

```python
X = DataFrame(data, columns=['production_budget_usd'])
y = DataFrame(data, columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.3)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

```python
regression = LinearRegression() #Creates  a 
#regression object that has the functionalities
# to perform a  linear regression
regression.fit(X, y)
```

Slope coefficient:

```python
regression.coef_    # theta_1 (gradient)--> 3.11
```

```python
#Intercept      This shows if the budget of a movie was 0$ the movie would lose 
# 7m dollars appx (Unrealistic) 
regression.intercept_
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, alpha=0.3)

# Adding the regression line here:
plt.plot(X, regression.predict(X), color='red', linewidth=3)

plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

```python
#Getting r square from Regression
regression.score(X, y)
```
