# Theoretical vs Experimental Probability



<p> Experimental probability and theoretical probability are two ways to figure out the chance of something happening.

Experimental probability is found by doing a test or experiment and counting how often an event happens. For example, if you flip a coin 10 times and it comes up heads 6 times, the experimental probability of heads is 6 out of 10 or 60%.

Theoretical probability uses math to calculate the chance of an event happening. It's based on the idea of what would happen if you did the same test over and over again an infinite number of times. For example, the theoretical probability of heads when flipping a coin is 50% because there are two possible outcomes (heads or tails) and each is equally likely.

In simple terms, experimental probability is based on what actually happened in a test, while theoretical probability is based on what should happen in an ideal world.</p>




$$P_{experimental}(Event A) = \frac{\text{No of times you observe A}}{\text{Total number of trials}}$$


$$P_{theoretical}(Event A) = \frac{\text{No of ways in which Event A can happen}}{\text{Total Number of ways any event can happen}}$$


```python
import random as rand
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
def probability_head(n):
    h =0
    t =0
    for i in range(n):
        x = rand.randint(0,1)
        if x==1:
            h+=1
        elif x==0:
            t+=1
        
    pHead=round(h/n,4)
    pTail=round(t/n,4)

    return pHead

```

```python


%matplotlib inline

n=int(input("No of trials"))
y_list =[]
x_list=range(1,n)
df = pd.DataFrame(x_list)
for i in range(1,n):
    z=probability_head(i)
    y_list.append(z)

# Make Data
x_1 = np.linspace(start=1, stop=10, num=10)
# print(x_1)
plt.xlim(0, n)
plt.ylim(0.3, 0.7)
plt.xlabel('Trials', fontsize=16)
plt.ylabel('Experimental Probability of Head', fontsize=16)
plt.scatter(x_list,y_list,color='pink', s=50, alpha=0.3)
plt.style.use('dark_background')
regr = LinearRegression()
regr.fit(df, y_list)
predicted_score=regr.predict(df)
plt.plot(x_list, predicted_score, color='teal', linewidth=3, alpha=1)
#test comment


plt.show()
```
